```python
from parser import *
import os
import pickle
```

```python
folder = './WEBPAGES_RAW/'

index = Index()

for sub_folder_id in range(75):
    for i in range(500):
        doc_id = sub_folder_id * 1000 + i
        fp = folder + '{}/{}'.format(sub_folder_id,i)
        
        try:
            str_ = parse(fp)
            token_dict = insert_token_dict(tokenize(str_,3),dict_={})
            for token,token_freq in token_dict.items():
                index.add(doc_id,token,token_freq)
                
            print(doc_id,fp)
        except OSError:
            print(fp, 'not found')
            break
```

```python
with open('index_file','w') as index_file:
    for token,posting_list in index.dict_.items():
        index_file.write(token)
        index_file.write('\t')
        index_file.write(str(posting_list))
        index_file.write('\n')
```

```python
index.search('noncommercial')
```

```python
with open('index_dump', 'wb') as handle:
    pickle.dump(index, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
